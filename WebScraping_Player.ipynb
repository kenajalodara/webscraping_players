```python
from selenium.webdriver import Chrome
import time
from bs4 import BeautifulSoup
import requests
```

```python
url = r'''http://mlb.mlb.com/stats/sortable.jsp#elem=%5Bobject+Object%5D&tab_level=child&click_text=Sortable+Player+hitting&game_type='R'&season=2019&season_type=ANY&league_code=MLB&sectionType=sp&statType=hitting&page=1&ts=1563313239622&sortColumn=ab&sortOrder='desc'&extended=0'''
```

```python
browser = Chrome()
browser.get(url)
time.sleep(5)
```

```python
html = browser.page_source
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
table_selector = ('div#datagrid table')
table = soup.select_one(table_selector)
```

```python
rows = table.select('tr')
```

```python
header_row = rows[0]
```

```python
headings = [th.text for th in header_row.select('th')]
```

```python
headings
```

```python
data_rows = rows[1:]
```

```python
data = []
for row in data_rows:
    data.append([td.text for td in row.select('td')])
  
```
